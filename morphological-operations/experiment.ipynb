```python
import cv2 as cv
from matplotlib import pyplot as plt
import numpy as np
```

```python
def plot_images(original, effected, effect_type):
    fig, axes = plt.subplots(1, 2, figsize=(15, 5))
    
    axes[0].imshow(original, cmap='gray')
    axes[0].set_title('Original Image')
    axes[0].axis('off')
    
    axes[1].imshow(effected, cmap='gray')
    axes[1].set_title(effect_type)
    axes[1].axis('off')
    
    plt.tight_layout()
    plt.show()

img_dir = f"/Users/lucky/projects/OpenCV-Experiments/data/Images/coin.jpg"
img = cv.imread(img_dir)
img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

plt.imshow(img, cmap='gray')
```

## Erosion Process

```python
# erosion and dilation
def apply_erosion(image, kernel_size=(5, 5)):
    kernel = np.ones(kernel_size, np.uint8)
    return cv.erode(image, kernel, iterations=1)

# Apply erosion and dilation
eroded_img = apply_erosion(img, kernel_size=(2, 2))
# Plot the results
plot_images(img, eroded_img, 'Eroded Image')
```

## Dilation Process

```python
def apply_dilation(image, kernel_size=(5, 5)):
    kernel = np.ones(kernel_size, np.uint8)
    return cv.dilate(image, kernel, iterations=1)

dilated_img = apply_dilation(img, kernel_size=(2, 2))
plot_images(img, dilated_img, 'Dilated Image')
```

## Opening Process

```python
def morph_opening(image, kernel_size=(5, 5)):
    kernel = np.ones(kernel_size, np.uint8)
    return cv.morphologyEx(image, cv.MORPH_OPEN, kernel)

opening_img = morph_opening(img, kernel_size=(8, 8))
plot_images(img, opening_img, 'Opening Image')
```

## Closing Process

```python
def morph_opening(image, kernel_size=(5, 5)):
    kernel = np.ones(kernel_size, np.uint8)
    return cv.morphologyEx(image, cv.MORPH_OPEN, kernel)

opening_img = morph_opening(img, kernel_size=(3, 3))
plot_images(img, opening_img, 'Opening Image')
```

## Morphological Gradient

```python
def gradient(image, kernel_size=(5, 5)):
    kernel = np.ones(kernel_size, np.uint8)
    return cv.morphologyEx(image, cv.MORPH_GRADIENT, kernel)

gradient_img = gradient(img, kernel_size=(2, 2))
plot_images(img, gradient_img, 'Morphological Gradient')
```

## Top Hat

```python
def top_hat(image, kernel_size=(5, 5)):
    kernel = np.ones(kernel_size, np.uint8)
    return cv.morphologyEx(image, cv.MORPH_TOPHAT, kernel)

top_hat_img = top_hat(img, kernel_size=(5, 5))
plot_images(img, top_hat_img, 'Top Hat')
```

## Balck Hat

```python
def black_hat(image, kernel_size=(5, 5)):
    kernel = np.ones(kernel_size, np.uint8)
    return cv.morphologyEx(image, cv.MORPH_BLACKHAT, kernel)

black_hat_img = black_hat(img, kernel_size=(5, 5))
plot_images(img, black_hat_img, 'Black Hat')
```
